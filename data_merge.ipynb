```python
def find_top10_ratio(data,topic):
  #topic 넣어주면 알아서 가져오기
  url = "https://openapi.naver.com/v1/datalab/search";
 
  for i in range(len(data[topic])):

    body = "{\"startDate\":\"2021-06-01\",\"endDate\":\"2021-06-30\",\"timeUnit\":\"date\",\"keywordGroups\":[{\"groupName\":\"CJ\",\"keywords\":[\"CJ\"]}],\"ages\":[\"1\",\"2\",\"3\",\"4\"]}";
    #if top10_3_1.columns[]
    
    for j in range(1,len(data.columns)):
        if data.columns[j]=='1':
          body = "{\"startDate\":\"2021-06-01\",\"endDate\":\"2021-06-30\",\"timeUnit\":\"date\",\"keywordGroups\":[{\"groupName\":\"CJ\",\"keywords\":[\"CJ\"]}],\"ages\":[\"1\",\"2\",\"3\",\"4\"]}";
        elif data.columns[j]=='2':
          body = "{\"startDate\":\"2021-06-01\",\"endDate\":\"2021-06-30\",\"timeUnit\":\"date\",\"keywordGroups\":[{\"groupName\":\"CJ\",\"keywords\":[\"CJ\"]}],\"ages\":[\"5\",\"6\",\"7\",\"8\",\"9\",\"10\"]}";
        else:
          body = "{\"startDate\":\"2021-06-01\",\"endDate\":\"2021-06-30\",\"timeUnit\":\"date\",\"keywordGroups\":[{\"groupName\":\"CJ\",\"keywords\":[\"CJ\"]}],\"ages\":[\"11\"]}";
    
    
    json_body=json.loads(body)

    for keyword in json_body['keywordGroups']:
      
        keyword['groupName']=data[topic][i]
        keyword['keywords']=list(data.loc[i:i,topic])

        body=json.dumps(json_body,ensure_ascii=False)
        body=str(body)

        request = urllib.request.Request(url)
        request.add_header("X-Naver-Client-Id",client_id)
        request.add_header("X-Naver-Client-Secret",client_secret)
        request.add_header("Content-Type","application/json")
        response = urllib.request.urlopen(request, data=body.encode("utf-8"))
        rescode = response.getcode()

        if(rescode==200):
            response_body = response.read()
            output_data=response_body.decode('utf-8')
        else:
            print("Error Code:" + rescode)

        result=json.loads(output_data)
        print(result)

        date = [a['period'] for a in result['results'][0]['data']]
        ratio = [a['ratio'] for a in result['results'][0]['data']]

        new=pd.DataFrame({'date':date, 'output':ratio})
            #print(new)
            #data=data.rename(columns={'output':'2020수능특강'})
        new=new.rename(columns={'output':data[topic][i]})
            #print(new)
            #new.to_csv("/content/drive/Shareddrives/2021 CJ 대한통운 공모전/data/crawling_data/top10/top10 세대별 ratio/3월/Allratio.csv",index=False)
        #j=np.arange(30)
        new.to_csv("/content/drive/Shareddrives/2021 CJ 대한통운 공모전/data/crawling_data/top10/top10 세대별 ratio/6월/6_"+data[topic][i]+".csv",index=False)
```
